```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
movies.shape
```

```python
movies.isna().sum()
movies.dropna(inplace=True)
movies.duplicated().sum()
movies.drop_duplicates()
```

```python
movies.iloc[0].genres #i need the name from genres
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
movies.head()
```

```python
#I want the first 3 main actors names
def top3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(top3)
```

```python
movies.head()
```

```python
print(movies['crew'][0])
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
#overview is a list
movies['overview'][0]
#so converting it into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#now concatenate all the columns which are a list and convert it to strings using join.
#firstly remove the spaces from keywords,cast and crew
def remove_space(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movie = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_movie.head()
```

```python
new_movie['tags'] = new_movie['tags'].apply(lambda x: " ".join(x))
new_movie.head()
```

```python
new_movie['tags'][0]
```

```python
new_movie['tags'][1]
#now i need to find the similarity score b/w these two movies using their tags for that i will convert them to vector. 
#We can get the same number of words similar but thats lengthy. so I use text vectorization.
#There are techniques-bagofwods,word2vec,tfidf. we will not use stop words like are,and,of, from
```

```python
#now we need to convert the movies to vectors. I will combine all the tags and find out the most 5000 common words used.
#      w1  w2  w3  w4
#m1    3   6
#m2    2   3
#m3    4   2
#3 times word1 has come in movie1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_movie['tags']).toarray()
```

```python
vector.shape
```

```python
feature_names= cv.get_feature_names_out()
```

```python
#for name in feature_names:
  #  print(name)
#we see that activity and activities are taken two times so we need to remove such occurrences for that we use stemming and for that 
#nltk 
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
#convert this to string again
    return "".join(y)
```

```python
new_movie['tags']=new_movie['tags'].apply(stem)
new_movie['tags']
```

```python
#now we have 4806 movies converted to vectors and each vector has 5000 words
```

```python
#I will find the angles between vectors. If the angle is less, then there is more similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
#in higher dimension euclidean is not reliableso we use cosine distance
similarity.shape
similarity[0]
```

```python
def recommend(movie):
    index = new_movie[new_movie['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movie.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```

```python
recommend('Gandhi')
```
